### 7/15/2026

## Overview

The SpaceX Launch Records Dashboard is an interactive data visualization application built using **Python, Dash, Pandas, and Plotly**. The dashboard allows users to explore SpaceX launch data and analyze launch success patterns based on launch sites, payload mass, and booster characteristics.

The goal of this dashboard is to provide an intuitive way to investigate historical SpaceX launch performance and identify relationships between payload size, launch locations, and mission success.

## Dashboard Features

The dashboard includes four interactive visualizations:

1. **Launch Success Distribution Across All Launch Sites**
   - Displays the total number of successful launches for each SpaceX launch site.
   - Allows users to compare launch site performance and identify locations with the highest number of successful missions.

2. **Launch Success Distribution for Individual Launch Sites**
   - Provides a detailed breakdown of successful and failed launches for a selected launch site.
   - Users can select a specific launch location from the dropdown menu to analyze its performance.

3. **Payload Mass vs Launch Success for All Sites**
   - Displays the relationship between payload mass and mission success across all launch sites.
   - Users can filter launches using the payload range slider.
   - Booster version categories are included to provide additional insight into how rocket configurations impacted mission outcomes.

4. **Payload Mass vs Launch Success for Individual Sites**
   - Allows users to analyze payload performance at a specific launch site.
   - Helps identify whether payload size influences launch success rates at different SpaceX facilities.

# Launch Success Distribution Across All Sites

## Description

This visualization presents the total successful launches across all SpaceX launch sites.

The pie chart aggregates successful missions by launch location, allowing users to quickly compare which launch sites have contributed the most successful missions.

## Purpose

This chart helps answer questions such as:

- Which launch sites have the highest number of successful launches?
- How are successful missions distributed across SpaceX facilities?
- Which locations have historically contributed the most to SpaceX launch operations?

## Interactive Features

- Automatically updates based on the complete SpaceX launch dataset.
- Each section of the pie chart represents a different launch site.
- Larger sections indicate a greater number of successful launches.

# Individual Launch Site Success Analysis

## Description

This visualization provides a detailed success and failure breakdown for a selected SpaceX launch site.

Users can select a specific launch site from the dropdown menu, and the pie chart updates to display the number of successful and unsuccessful missions for that location.

## Purpose

This chart allows users to evaluate the performance of individual launch locations.

It helps answer:

- How reliable has a specific launch site been?
- What percentage of missions from a location were successful?
- How do different launch sites compare in terms of mission outcomes?

## Interactive Features

- Dropdown menu for selecting launch sites.
- Dynamic chart updates based on user selection.
- Displays successful and failed launch counts.

# Payload Mass vs Launch Success Across All Sites

## Description

This scatter plot analyzes the relationship between payload mass and launch success across all SpaceX launch sites.

Each point represents an individual launch, with payload mass displayed on the x-axis and mission outcome represented on the y-axis.

The visualization also uses booster version categories to identify differences between rocket configurations.

## Purpose

This chart helps explore whether payload size has an impact on launch success.

It helps answer:

- Are heavier payloads associated with more launch failures?
- How do different booster versions perform?
- Are certain payload ranges more successful?

## Interactive Features

- Payload range slider allows users to filter launches by payload mass.
- Color coding separates different booster versions.
- Updates dynamically based on selected payload ranges.

# Payload Mass vs Launch Success for Individual Sites

## Description

This visualization focuses on the relationship between payload mass and launch success at a specific SpaceX launch site.

Users can select a launch site and analyze how payload size and booster versions relate to mission outcomes.

## Purpose

This chart provides a deeper analysis of individual launch facilities.

It helps answer:

- Does payload mass influence success rates at a particular site?
- How does a launch site's performance change with different payload sizes?
- Which booster versions have performed best at each location?

## Interactive Features

- Select a launch site using the dropdown menu.
- Adjust payload mass limits using the range slider.
- View individual launch outcomes and booster performance patterns.
